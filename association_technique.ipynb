## Installing and importing libraries

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

## Data Loading and Cleaning

```python
df = pd.read_excel('/content/Online Retail.xlsx')
df.head()
```

```python
df['Country'].unique()
```

```python
df = df[~df['Country'].str.contains('Unspecified')]
```

```python
df.isnull().sum()
```

```python
df.dropna(axis = 0, inplace = True)
```

```python
desc = df['Description'].unique()
```

```python
df['Description'] = df['Description'].astype(str)
df = df[~df['Description'].str.contains('wrong')]
```

```python
df.dropna(axis = 0, subset = ['InvoiceNo'], inplace = True)
df['InvoiceNo'] = df['InvoiceNo'].astype(str)
df = df[~df['InvoiceNo'].str.contains('C')]
df = df[~df['InvoiceNo'].str.contains('A')]

df.head()
```

```python
df.shape
```

```python
df['Description']= df['Description'].str.strip()
```

```python
df.head()
```

## Working with Complete dataset or Looking at the complete world's consumer data

```python
basket = (df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket.head()
```

```python
basket.shape
```

```python
basket.sum(axis = 0, skipna = True).sort_values(ascending = True).head(2200)
```

```python
basket = basket.iloc[:, :-2200]
```

```python
basket.shape
```

```python
def encode_units(x):
  if x<= 0:
    return 0
  elif x>=1:
    return 1

basket_set = basket.applymap(encode_units)
```

```python
basket_set
```

## Association techniques

```python
from mlxtend import frequent_patterns
import timeit
tic = timeit.default_timer()
freq_itemset = apriori(basket_set, min_support= .01, use_colnames=True)
rules = association_rules(freq_itemset, metric = 'lift', min_threshold=1)
toc = timeit.default_timer()
print(tic-toc)
```

```python
rules
```

```python
rules[(rules['lift']>= .5) & (rules['confidence'] >= .5)]
```

```python

from mlxtend.frequent_patterns import fpgrowth
```

```python
tic = timeit.default_timer()
freq_itemset = fpgrowth(basket_set, min_support= .01, use_colnames=True)
rules = association_rules(freq_itemset, metric = 'lift', min_threshold=1)
toc = timeit.default_timer()
print(tic-toc)
```

```python
rules
```

```python
rules[(rules['lift']>= .5) & (rules['confidence'] >= .5)]
```

## we can see there is good amount of time difference between the running time of FP-growth method and Apriori method and if we compare the rules they are almost similar to each other
## so we can conclude that Fp-growth is better than Apriori in case of time complexity
